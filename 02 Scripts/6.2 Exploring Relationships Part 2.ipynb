```python
# Refining education column:

df['mother_ed'].value_counts()
```

```python
# Consolidating mother education categories for better categorical visuals.

df.loc[df['mother_ed'] == "high school grad/GED", 'mother_ed'] = "high school/GED or less"
df.loc[df['mother_ed'] == "grade 9-12 no diploma", 'mother_ed'] = "high school/GED or less"
df.loc[df['mother_ed'] == "8th grade or less", 'mother_ed'] = "high school/GED or less" 
df.loc[df['mother_ed'] == "some college but no degree", 'mother_ed'] = "associate degree or some college" 
df.loc[df['mother_ed'] == "associate degree", 'mother_ed'] = "associate degree or some college"
```

```python
df['mother_ed'].value_counts()
```

```python
# Creating birth weight alert categories. These categories are considered low birth weight (LBW), normal birth weight
# (NBW), high birth weight (HBW), very low birth weight (VLBW), and extremely low birth weight (ELBW). 

df.loc[(df['birth_wt_avg'] < 1000), 'birth_wt_alert'] = 'ELBW'
df.loc[(df['birth_wt_avg'] >= 1000) & (df['birth_wt_avg'] < 1500), 'birth_wt_alert'] = 'VLBW'
df.loc[(df['birth_wt_avg'] >= 1500) & (df['birth_wt_avg'] < 2500), 'birth_wt_alert'] = 'LBW'
df.loc[(df['birth_wt_avg'] >= 2500) & (df['birth_wt_avg'] <= 4000), 'birth_wt_alert'] = 'NBW'
df.loc[(df['birth_wt_avg'] > 4000), 'birth_wt_alert'] = 'HBW'
```

```python
df.head()
```

```python
df.tail()
```

```python
df['birth_wt_alert'].value_counts(dropna=False)
```

```python
df.shape
```

```python
# Gestational age categories from March of Dimes and WHO. 

# Creating gestational age alert categories. These categories are considered not preterm, preterm, very preterm, and
# extremely preterm. 

df.loc[(df['gest_age_avg'] >= 42), 'gest_age_alert'] = 'postterm'
df.loc[(df['gest_age_avg'] >= 41) & (df['gest_age_avg'] < 42), 'gest_age_alert'] = 'late-term'
df.loc[(df['gest_age_avg'] >= 37) & (df['gest_age_avg'] < 41), 'gest_age_alert'] = 'not preterm'
df.loc[(df['gest_age_avg'] >= 32) & (df['gest_age_avg'] < 37), 'gest_age_alert'] = 'preterm'
df.loc[(df['gest_age_avg'] >= 28) & (df['gest_age_avg'] < 32), 'gest_age_alert'] = 'very preterm'
df.loc[(df['gest_age_avg'] < 28), 'gest_age_alert'] = 'extremely preterm'
```

```python
df.head()
```

```python
df.tail()
```

```python
df.shape
```

```python
# Creating mother age category column based on quartiles. This technique is being used instead of an established young/old 
# mother age ranking system because we are dealing with averages and not individual mother ages. Because of this,
# categories such as really young (< 18) and older mothers (> 40) do not exist in the data. If a category has an average in 
# the 30s, for example, it is likely to contain individuals who are much older than that to achieve that average.

df['mother_age_avg'].quantile([0.2, 0.4, 0.6, 0.8])
```

```python
df.loc[(df['mother_age_avg'] < 26), 'mother_age_range'] = '< 26'

df.loc[(df['mother_age_avg'] >= 26) & (df['mother_age_avg'] < 28), 'mother_age_range'] = '26-28'

df.loc[(df['mother_age_avg'] >= 28) & (df['mother_age_avg'] < 30), 'mother_age_range'] = '28-30'

df.loc[(df['mother_age_avg'] >= 30) & (df['mother_age_avg'] <= 32), 'mother_age_range'] = '30-32'

df.loc[(df['mother_age_avg'] > 32), 'mother_age_range'] = '> 32'
```

```python
df['mother_age_range'].value_counts(dropna=False)
```

```python
# Making a birth weight category based on quantiles instead of established alert categories.This technique is being used 
# instead of an established birth weight categories because we are dealing with averages and not individual mother ages. 
# Because of this, a low or high birth weight alert category may not draw enough attention to records that are lower or
# higher than they should be. Because birth weights are average birth weights by data grain, they should have more modest
# thresholds than alert categories. 

df['birth_wt_alert'].value_counts(dropna=False)
```

```python
df['birth_wt_alert'].isnull().sum()
```

```python
# This will allow the creation of 5 birth weight categories, similar to the alerts.

df['birth_wt_avg'].quantile([0.2, 0.4, 0.6, 0.8])
```

```python
# Creating birth weight category column based on quantiles set above.

df.loc[(df['birth_wt_avg'] < 3108.951953), 'birth_wt_cat'] = 'lowest'

df.loc[(df['birth_wt_avg'] >= 3108.951953) & (df['birth_wt_avg'] < 3202.243994), 'birth_wt_cat'] = 'low'

df.loc[(df['birth_wt_avg'] >= 3202.243994) & (df['birth_wt_avg'] < 3270.540039), 'birth_wt_cat'] = 'average'

df.loc[(df['birth_wt_avg'] >= 3270.540039) & (df['birth_wt_avg'] <= 3342.379883), 'birth_wt_cat'] = 'high'

df.loc[(df['birth_wt_avg'] > 3342.379883), 'birth_wt_cat'] = 'highest'
```

```python
# Now, we will create gestational categories based on quantiles just as we did for mother age and birth weight categories.

df['gest_age_alert'].value_counts(dropna=False)
```

```python
# This will allow the creation of 5 gestational age categories, similar to the alerts.

df['gest_age_avg'].quantile([0.2, 0.4, 0.6, 0.8])
```

```python
# Creating gestational age category column based on quantiles set above.

df.loc[(df['gest_age_avg'] < 38), 'gest_age_cat'] = 'youngest'

df.loc[(df['gest_age_avg'] >= 38) & (df['gest_age_avg'] < 38.369999), 'gest_age_cat'] = 'young'

df.loc[(df['gest_age_avg'] >= 38.369999) & (df['gest_age_avg'] < 38.610001), 'gest_age_cat'] = 'average'

df.loc[(df['gest_age_avg'] >= 38.610001) & (df['gest_age_avg'] <= 38.880001), 'gest_age_cat'] = 'old'

df.loc[(df['gest_age_avg'] > 38.880001), 'gest_age_cat'] = 'oldest'
```

### 4. Exploring relationships

#### a) Finishing up pairplots

```python
sub = df[['prenatal_start','births','birth_wt_avg','mother_age_avg','gest_age_avg','birth_wt_cat']]
```

```python
from matplotlib import rcParams
rcParams['axes.titlepad'] = 20 
```

```python
hue_order = ['lowest','low','average','high','highest']
```

```python
prplt_bwcat = sns.pairplot(sub,hue='birth_wt_cat',palette='coolwarm', hue_order=hue_order)
```

#### Here, we see some of the same trends as with the heatmap. Older mothers appear to give birth to infants of larger birth weights and somewhat longer gestational age. We also see the weak realtionship between earlier prenatal start month and more lower than average birth weights.

```python
prplt_bwcat.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'prplt_bwcat.png'), bbox_inches='tight')
```

```python
sub_2 = df[['prenatal_start','births','birth_wt_avg','mother_age_avg','gest_age_avg','birth_wt_alert']]
```

```python
# Here, you can see that even though older women are inclined to give birth to heavier infants and have longer gestational
# ages on average, these older women still give birth to infants in normal birth weights and gestation ranges.

hue_order_2 = ['ELBW','VLBW','LBW','NBW','HBW']
prplt_bwalert = sns.pairplot(sub_2,hue='birth_wt_alert',palette='coolwarm', hue_order=hue_order_2)

# Cleared output for optimized script performance.
```

```python
prplt_bwalert.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'prplt_bwalert.png'), bbox_inches='tight')
```

```python
corr_age_wt_bwcat = sns.lmplot(x='mother_age_avg',y='birth_wt_avg',data=sub,hue='birth_wt_cat',palette='coolwarm', hue_order=hue_order)
plt.title('Mother Age & Birth Weight Correlation by Weight Category', fontsize=10)
plt.xlabel('Average Mother Age')
plt.ylabel('Average Birth Weight')
```

```python
corr_age_wt_bwcat.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'corr_age_wt_bwcat.png'), bbox_inches='tight')
```

#### Here, we see a larger, clearer picture of one of our most intriguing correlations, that of mother age and birth weight. We can see that mothers in their 30s give birth to lower birth weight infants than other age groups (which are still likely mostly considered normal weight according to medically established birth weight categories). Also, you can see that heavier infants are slightly skewed at the top, meaning older women tend to give birth to larger babies.

#### b) Categorical plots

```python
# Categorical plot showing prenatal care and birth weights.

sns.set(style="ticks")
prental_wt_cat = sns.catplot(x="prenatal_start", y="birth_wt_avg", hue="birth_wt_cat", data=df, palette='coolwarm', hue_order=hue_order)
plt.title('Birth Weight by Prenatal Care Start Month', fontsize=12)
plt.xlabel('Prental Care Start Month', fontsize=10)
plt.ylabel('Birth Weight Average in Grams', fontsize=10) 
```

```python
prental_wt_cat.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'prental_wt_cat.png'), bbox_inches='tight')
```

#### Women who start prenatal care sooner are inclined to disproportionally deliver more lower weight infants, with those who start prenantal care later, have fewer lower birth weight infants, the opposite of what one might expect. However, 0 represents "No prenatal care," so this will later be relabeled to 10 to more accurately show the birth weight category of this group. Is this pattern due to more women starting care around the 1-4 LMP range, so they are having more lower weight babies, or is there another reason?

```python
# Categorical plot showing prenatal care and gestational age.

sns.set(style="ticks")
prental_gest_cat = sns.catplot(x="prenatal_start", y="gest_age_avg", hue="gest_age_cat", data=df, palette='coolwarm', hue_order=hue_order_3)
plt.title('Gestational Age by Prenatal Care Start Month', fontsize=12)
plt.xlabel('Prental Care Start Month', fontsize=10)
plt.ylabel('Gestational Age Average in Weeks', fontsize=10) 
```

#### Thoughts: Do the earlier months have more lower gestational ages because more women start prenatal care earlier (more issues like low birth weight or earlier gestational age are present) because there are more women starting care ealier, or are more women miscarrying at this gestional age? 

```python
prental_gest_cat.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'prental_gest_cat.png'), bbox_inches='tight')
```

```python
# Categorical plot showing moother education and birth weight.

sns.set(style="ticks")
education_wt_cat = sns.catplot(x="mother_ed", y="birth_wt_avg", hue="birth_wt_cat", data=df, palette='coolwarm', hue_order=hue_order)
plt.title("Birth Weight by Mother's Education Level", fontsize=12)
plt.xlabel('Mother Education', fontsize=12)
plt.ylabel('Birth Weight Average in Grams', fontsize=12) 
education_wt_cat.set_xticklabels(rotation=90)

# Can also use this to rotate xtick labels: plt.xticks(rotation = 25)

```

```python
education_wt_cat.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'education_wt_cat.png'), bbox_inches='tight')
```

#### Thoughts: Women with bachelor's, master's, and Ph.D. or professional degrees, seem to have the same proportion of lower, normal, and higher birth weights. Is there a reason why women with high school education levels or some college have greater levels of lower and greater birth weight infacts? Remember, these are averages, and even though these are lower and higher average, most do not fall into established low or high birth weight categories.

```python
# Categorical plot showing mother education and gestational age.

sns.set(style="ticks")
education_gest_cat = sns.catplot(x="mother_ed", y="gest_age_avg", hue="gest_age_cat", data=df, palette='coolwarm', hue_order=hue_order_3)
plt.title("Gestational Age by Mother's Education Level", fontsize=12)
plt.xlabel('Mother Education', fontsize=12)
plt.ylabel('Gestational Age Average in Weeks', fontsize=12) 
plt.xticks(rotation = 90)
```

#### Thoughts: There is a similar pattern here with mother education and gestational age as there is with mother education and birth weight. Again, these gestational categories are based on population quantiles and not established low and high gestational ages, meaning young and old here does not necessarily mean (but it can) dangerously low or or high gestational ages.

```python
education_gest_cat.figure.savefig(os.path.join(path, '04 Analysis', 'Visualizations', 'education_gest_cat.png'), bbox_inches='tight')
```
